```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Создаем данные
date = pd.date_range(start='2021-09-01', freq='D', periods=8)
temperature = [23, 17, 17, 16, 15, 14, 17, 20]
data = pd.DataFrame({'Дата': date, 'Температура': temperature})

# Настраиваем стиль Seaborn
sns.set(style='whitegrid')

# Строим график
plt.figure(figsize=(8, 4))  # Устанавливаем размер графика
sns.lineplot(x='Дата', y='Температура', data=data, marker='o', color='b')

# Задаем название графика
plt.title('График температуры')

# Показываем график
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Создаем небольшой DataFrame с данными
data = {
    'sepal_length': [5.1, 4.9, 5.7, 6.0, 5.6],
    'sepal_width': [3.5, 3.0, 3.8, 3.2, 3.9]
}

# Добавляем данные в ваш DataFrame iris_data
iris_data = iris_data.append(pd.DataFrame(data), ignore_index=True)

# Теперь мы можем построить точечный график
iris_data.plot.scatter(x='sepal_length', y='sepal_width')

# Устанавливаем подписи для осей
plt.xlabel('Длина чашелистника (см)')
plt.ylabel('Ширина чашелистника (см)')

# Отображаем график
plt.show()
```
